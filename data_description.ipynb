🔍 Enfoque del Análisis Exploratorio
1. Overview del Dataset
Dimensiones de las tablas (train/test, oot, master, históricas)

Proporción de clases en la variable objetivo (balance/desbalance)

2. Análisis Univariado
Distribuciones de las variables numéricas (e.g., histogramas, boxplots)

Proporciones de variables categóricas

Identificación de valores extremos

3. Análisis Bivariado
Relación de cada variable con la variable objetivo

Numéricas: violinplot, boxplot, kdeplot, pointplot

Cálculo de métricas estadísticas (correlación, chi-cuadrado, IV)

4. Análisis Multivariado
Correlación entre variables numéricas

PCA para visualizar agrupaciones si es útil

Detección de multicolinealidad (VIF)

5. Análisis Temporal (si aplica)
Evolución de variables clave por mes o corte de tiempo (si tienes variable temporal)

6. Data Drift entre Train y OOT
Comparación de distribuciones

Cambio en proporciones y estadísticas de las variables entre train y oot
